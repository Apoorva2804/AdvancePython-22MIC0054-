 #Basic matplotlib function

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])#plot is a versatile function, and will take an arbitrary number of arguments. For example, to plot x versus y
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')#optional third argument which is the format string that indicates the color and line type of the plot
plt.axis((0, 6, 0, 20))
plt.show()
```

```python
#plotting several lines with different format styles in one function call using arrays
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

#all sequences are converted to numpy arrays internally

```python
#Matplotlib allows to provide such an object with the data keyword argument
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

# IMAGE MANIPULATION

```python
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
```

Loading the image

```python
img = np.asarray(Image.open('D:\\Apoo\\assignments\\Picasso-Bug.jpg'))
print(repr(img))
```

# Plotting numpy arrays as images

```python
imgplot = plt.imshow(img)
```

# Applying pseudocolor schemes to image plots

```python
lum_img = img[:, :, 0]
plt.imshow(lum_img)
```

# Hot colour scheme

```python
plt.imshow(lum_img, cmap="hot")
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

# Color scale reference

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

# Histogram for the image

```python
plt.hist(lum_img.ravel(), bins=range(256), fc='k', ec='k')
```

The histogram you plotted represents the grayscale intensity distribution of your bug image. The x-axis shows pixel intensity values ranging from 0 (black) to 255 (white), while the y-axis shows how many pixels in the image fall into each intensity range. In your case, the histogram has noticeable peaks in the mid-tone region around 90130, which correspond to the natural textures on the branch and parts of the bugs body. There is also a strong spike near 255, indicating a significant number of very bright pixels from highlights and lighter background areas. A smaller bump appears in the lower range around 2040, reflecting the darker shadows such as under the branch and the bugs legs. Overall, the histogram shows that the image is dominated by mid-tones and bright regions, with fewer very dark areas.

# effectively "zoom in on" part of the histogram,we do this by setting clim, the colormap limits.

```python
plt.imshow(lum_img, clim=(0, 175))
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0, 175)
```
